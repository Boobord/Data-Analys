```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix


Heart_disease=pd.read_csv('D:/power points/Datascience workshop/datascience data&codes/Heart Disease.csv')
x=Heart_disease.iloc[:,0:279]
y=Heart_disease['class']
Heart_disease.shape
```

```python
Heart_disease
```

```python
print(x)
```

```python
print(y)
```

```python
Heart_disease['class'].replace('Arrhythmia',1, inplace= True)
Heart_disease['class'].replace('Normal',0, inplace= True)
```

```python
Heart_disease
```

```python
#can not find the nan because python does not know "?" as a missing value
# to solve this problem, we replace "?" with "nan" that is known by python as missing value
H =x.isnull().sum()
print(H)
```

```python
print(*H)
```

```python
#replacing "?" with "nan"
x.replace('?',np.nan,inplace=True)
```

```python
#know python know the missing values
H = x.isnull().sum()
print(*H)
```

```python
# columns that have missing values
nan_values = x.isna()
nan_column= nan_values.any()
columns_with_nan=x.columns[nan_column].tolist()
print('columns_with_nan:',columns_with_nan)
```

```python
from sklearn.impute import SimpleImputer

for f in columns_with_nan:
    imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
    imputer.fit(x[f].values.reshape(-1,1))
    x[f]=imputer.transform(x[f].values.reshape(-1,1))
```

```python
d = Heart_disease.isnull().sum()
print(*d)
```

# Support vector machine (SVM) with missing value

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
predictions=svc.predict(x_test)
predictions
```

```python
print (confusion_matrix(y_test,predictions,labels=[1,0]))
```

```python
#cm=Confusion Matrix
cm=np.array(confusion_matrix(y_test,predictions,labels=[1,0]))
confusion=pd.DataFrame(cm, index=['is_Arrhythmia','isNOT_Arrhythmia'],columns=['predict_Arrhythmia',
                                                                               'predict_NOTArrhythmia'])

confusion
```

```python
#accuracy of SVM using confusion matrix for evaluation
from sklearn import metrics
print ("Accuracy of SVM:", metrics.accuracy_score(y_test,predictions))
print(classification_report(y_test,predictions))
```

```python
#accuracy of SVM after using k-fold for evaluation
from sklearn.model_selection import KFold
kf = KFold(n_splits = 5, shuffle = True)
kf_svclassifier = SVC()
scores = []
for i in range(5):
    result = next(kf.split(x), None)
    #print(result)
    
    xxx_train = x.iloc[result[0]]
    #print(xxx_train)
    yyy_train = y.iloc[result[0]]
    #print("yyyy_train",yyy_train)
    
    xxx_test=x.iloc[result[1]]
    yyy_test = y.iloc[result[1]]
    
    model = kf_svclassifier.fit(xxx_train,yyy_train)
    predictions = kf_svclassifier.predict(xxx_test)
    scores.append(model.score(xxx_test,yyy_test))
    
    print(confusion_matrix(yyy_test,predictions))
    print(classification_report(yyy_test,predictions))
    print('Scores from each Iteration: ', scores)
    Accuracy = np.mean(scores)
print('Average K-Fold Score :' , np.mean(scores))
```

# Naive Bayesian Algorithm (Dimension Reduction)

```python
from sklearn.decomposition import PCA
```

```python
dim_r=PCA(n_components=10)
dim_r.fit(x)
dim_red= dim_r.transform(x)
print(dim_red)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(dim_red,y,test_size=0.2)
GNB=GaussianNB()
```

```python
GNB.fit(x_train,y_train)
```

```python
predictions=GNB.predict(x_test)
predictions
```

```python
print (confusion_matrix(y_test,predictions,labels=[1,0]))
```

```python
from sklearn import metrics
print ("Accuracy  of naive bayesian after using PCA:", metrics.accuracy_score(y_test,predictions))
print(classification_report(y_test,predictions))
```
