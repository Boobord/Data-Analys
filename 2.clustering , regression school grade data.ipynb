# Linear Regression Algorithms

```python

import pandas as pd
import numpy as np
from sklearn import preprocessing
school_grade=pd.read_csv('school_grades_dataset.csv')


school_grade['sex'].replace('F',1, inplace= True)
school_grade['sex'].replace('M',0, inplace= True)

school_grade['paid'].replace('yes',1, inplace= True)
school_grade['paid'].replace('no',0, inplace= True)

school_grade['internet'].replace('yes',1, inplace= True)
school_grade['internet'].replace('no',0, inplace= True)
#print(school_grade)


x=school_grade.iloc[:,0:18]
y=school_grade['G3']

#print(school_grade)
school_grade.shape
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.1, random_state=42)
```

```python
#x_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
y_pred
```

```python
from sklearn import metrics
mse=metrics.mean_squared_error(y_test,y_pred)
mse
```

# Polynomiyal regresion algorithm

```python
from sklearn.preprocessing import PolynomialFeatures 
  
poly_reg = PolynomialFeatures(degree = 4) 
x_poly = poly_reg.fit_transform(x_train) 
pol_reg = LinearRegression()
pol_reg.fit(x_poly, y_train)
y_predict=pol_reg.predict(poly_reg.fit_transform(x_test))
y_predict
```

```python
mse=metrics.mean_squared_error(y_test,y_predict)
mse
```

```python
kmn=KMeans(n_clusters=3)
kmn.fit(iris.data)
label_iris=kmn.predict(iris.data)
#print(labels)

print(*label_iris)
```

```python
centroid_iris = kmn.cluster_centers_
centroid_iris
```

# 2. Hierarchical Algorithm

```python
from sklearn import datasets
iris=datasets.load_iris()
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
```

```python
hierarchical1 = linkage(iris.data, method='complete')
dendrogram(hierarchical1)
plt.show()
```

```python
label=fcluster(hierarchical1,3,criterion='distance')
print(*label)
```
